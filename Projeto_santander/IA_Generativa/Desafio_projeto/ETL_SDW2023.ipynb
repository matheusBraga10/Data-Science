```python
import sqlite3 as conector
```

```python
try:
  # abertura da conexão
  conexao = conector.connect('/content/drive/MyDrive/Data Science/Santander - DIO/Desafio-ETL/SDW2023.db')
# aquisição de um cursor
  cursor = conexao.cursor()

  cria_tabela_01 = '''
  CREATE TABLE usuario
  (
  id INT(11) NOT NULL,
  nome VARCHAR(30) NOT NULL,
  PRIMARY KEY (nome)
  );
  '''
  cria_tabela_0 = '''
  CREATE TABLE card
  (
  number_card VARCHAR(20) NOT NULL,
  limit_card INT(20) NOT NULL,
  PRIMARY KEY (number_card)
  );
  '''
  cria_tabela_03 = '''
  CREATE TABLE feactures
  (
  id INT(2),
  icon VARCHAR(200),
  description VARCHAR(200),
  PRIMARY KEY (id)
  );
  '''
  cria_tabela_04 = '''
  CREATE TABLE news
  (
  id INT(2),
  icon VARCHAR(200),
  description VARCHAR(200),
  PRIMARY KEY (id)
  );
  '''
  cria_tabela_05 = '''
  CREATE TABLE account
  (
  id INT(2),
  nome VARCHAR(30),
  number_card VARCHAR(20),
  number_account VARCHAR(7) NOT NULL,
  agency VARCHAR(4) NOT NULL,
  balance INT(10),
  limit_account INT(10) NOT NULL,
  FOREIGN KEY (nome) REFERENCES usuario (nome),
  FOREIGN KEY (number_card) REFERENCES card (number_card),
  FOREIGN KEY (id) REFERENCES feactures (id),
  FOREIGN KEY (id) REFERENCES news (id),
  PRIMARY KEY (number_account)
  );
  '''

  # execução do comando: SELECT, INSERT, CREATE...
  cursor.execute(cria_tabela_01)
  cursor.execute(cria_tabela_02)
  cursor.execute(cria_tabela_03)
  cursor.execute(cria_tabela_04)
  cursor.execute(cria_tabela_05)

# Efetivação do comando
  conexao.commit()


except conector.DatabaseError as erro:
  print('Erro no banco de dados', erro)
finally:
  # fechamento das conexoes
  cursor.close()
  conexao.close()
```
